```python
import random

scores = []

while len(scores) < 20:
    score = random.randint(0, 6)
    
    if score == 0:
        break
    
    if score == 6:
        continue
    
    scores.append(score)

print(scores)

```

```python
# Cell 2 - Data Validation
print("Total valid responses:", len(scores))
```

```python
# Cell 3 - Average Score
average = round(sum(scores) / len(scores), 2)
print("Average score:", average)
```

```python
# Cell 4 - Min / Max Analysis
print("Min score:", min(scores))
print("Max score:", max(scores))
```

```python
# Cell 5 - Satisfaction Analysis
satisfied = 0
neutral = 0
unsatisfied = 0

for score in scores:
    if score >= 4:
        satisfied += 1
    elif score == 3:
        neutral += 1
    else:
        unsatisfied += 1

print("Satisfied:", satisfied)
print("Neutral:", neutral)
print("Unsatisfied:", unsatisfied)
```

```python
# Cell 6 - Early Warning System (Break)
consecutive_ones = 0

for score in scores:
    if score == 1:
        consecutive_ones += 1
        if consecutive_ones == 3:
            print("⚠️ Critical dissatisfaction detected!")
            break
    else:
        consecutive_ones = 0
```

```python
# Final Cell - General Summary
print("Total valid responses:", len(scores))
print("Average score:", round(sum(scores) / len(scores), 2))
print("Min score:", min(scores))
print("Max score:", max(scores))
print()
print("Satisfied:", satisfied)
print("Neutral:", neutral)
print("Unsatisfied:", unsatisfied)
```
